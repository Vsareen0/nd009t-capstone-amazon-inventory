## Data Preparation


```python
import json
import random
import os

def create_splits(data_dir):
    with open(data_dir, 'r') as f:
        data = json.load(f)

    splits = {}
    for key, values in data.items():
        random.shuffle(values)
        split_indexes = {
            'train': int(0.8 * len(values)),
            'valid': int(0.1 * len(values)),
            'test': int(0.1 * len(values))
        }
        splits[key] = {
            'train': values[:split_indexes['train']],
            'valid': values[split_indexes['train']:split_indexes['train']+split_indexes['valid']],
            'test': values[split_indexes['train']+split_indexes['valid']:]
        }
        print(f"{key} has {len(splits[key]['train'])} training samples, {len(splits[key]['valid'])} validation samples, and {len(splits[key]['test'])} testing samples.")

    for split_name in ['train', 'valid', 'test']:
        with open(f'{split_name}.json', 'w') as f:
            json.dump({key: values[split_name] for key, values in splits.items()}, f)


data_dir = 'file_list.json'
create_splits(data_dir)
```

```python
import os
import json
import boto3

def download_and_arrange_data(folder):
    s3_client = boto3.client('s3')

    with open(f"{folder}.json", 'r') as f:
        d=json.load(f)

    for k, v in d.items():
        print(f"Downloading Images with {k} objects")
        directory=os.path.join(folder, k)
        if not os.path.exists(directory):
            os.makedirs(directory)
        for file_path in tqdm(v):
            file_name=os.path.basename(file_path).split('.')[0]+'.jpg'
            s3_client.download_file('aft-vbi-pds', os.path.join('bin-images', file_name),
                             os.path.join(directory, file_name))
```

```python
download_and_arrange_data('train')
download_and_arrange_data('test')
download_and_arrange_data('valid')
```

```python
#TODO: Perform any data cleaning or data preprocessing

import numpy as np
import matplotlib.pyplot as plt

with open(f"file_list.json", 'r') as f:
        d=json.load(f)
        
num_objects = [('Objects: ' + num) for num in d.keys()]
values = [len(v) for v in d.values()]

fig, ax = plt.subplots(figsize=(8, 8))

ax.bar(num_objects, values, color='b', width=0.4)
for bar in ax.patches:
    ax.annotate(format(bar.get_height() , '.2f'), 
                (bar.get_x() + bar.get_width()  / 2, bar.get_height()),
                ha='center',
                va='center',
                size=8,
                xytext=(0, 8),
                textcoords='offset points')
    
plt.xlabel('Number of objects in bin')
plt.ylabel('Values')
plt.title('Class wise distribution of number of objects in bin')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import json

with open(f"train.json", 'r') as f:
        trainData=json.load(f)
with open(f"test.json", 'r') as f:
        testData=json.load(f)
with open(f"valid.json", 'r') as f:
        validData=json.load(f)
        
trainValues = [len(v) for v in trainData.values()]
testValues = [len(v) for v in testData.values()]
validValues = [len(v) for v in validData.values()]

datasetBin = [('Objects: ' + num) for num in trainData.keys()]
bin_means = {
    'train': trainValues,
    'test': testValues,
    'valid': validValues,
}

x = np.arange(len(datasetBin))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in bin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Values')
ax.set_title('Class wise distribution of number of objects in bin')
ax.set_xticks(x + width, datasetBin)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 2500)

plt.show()
```

## Model Training

### Hyperparameter Tuning
- We are taking a continous space for learning rate and Categorical space for batch size with discrete values.


```python
#TODO: Declare your model training hyperparameter.
#NOTE: You do not need to do hyperparameter tuning. You can use fixed hyperparameter values
hyperparameter_ranges = {
    "learning-rate": ContinuousParameter(0.001, 0.01),
    "batch-size": CategoricalParameter([16, 32, 64, 128, 256])
}

objective_metric_name = "Test Loss"
objective_type = "Minimize"
metric_definitions = [{"Name": "Test Loss", "Regex": "Testing Loss: ([0-9\\.]+)"}]
```

```python
#TODO: Create your training estimator
estimator = PyTorch(
    entry_point="hpo.py",
    role=role,
    py_version='py36',
    framework_version="1.8",
    instance_count=1,
    instance_type="ml.g4dn.xlarge"
)

tuner = HyperparameterTuner(
    estimator,
    objective_metric_name,
    hyperparameter_ranges,
    metric_definitions,
    max_jobs=5,
    max_parallel_jobs=2,
    objective_type=objective_type,
)
```

```python
# TODO: Fit your estimator
tuner.fit({
    'training': train,
    'test': test,
    'valid': valid
}, wait=True) # Remember to include your data channels
```

```python
# Find the best hyperparameters
best_estimator = tuner.best_estimator()

best_estimator.hyperparameters()
```

```python
def get_data(trial, tname, mode):
    tensor = trial.tensor(tname)
    steps = tensor.steps(mode=mode)
    vals = []
    print(f"[INFO] Getting data in {len(steps)} steps")
    for s in steps:
        vals.append(tensor.value(s, mode=mode))
        print(".", end="")
    return steps, vals
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import host_subplot

def plot_tensor(trial, tensor_name):

    steps_train, vals_train = get_data(trial, tensor_name, mode=ModeKeys.TRAIN)
    print("loaded TRAIN data")
    steps_eval, vals_eval = get_data(trial, tensor_name, mode=ModeKeys.EVAL)
    print("loaded EVAL data")

    fig = plt.figure(figsize=(20, 7))
    host = host_subplot(111)

    par = host.twiny()

    host.set_xlabel("Steps (TRAIN)")
    par.set_xlabel("Steps (EVAL)")
    host.set_ylabel(tensor_name)

    (p1,) = host.plot(steps_train, vals_train, label=tensor_name)
    print("completed TRAIN plot")
    (p2,) = par.plot(steps_eval, vals_eval, label="val_" + tensor_name)
    print("completed EVAL plot")
    leg = plt.legend()

    host.xaxis.get_label().set_color(p1.get_color())
    leg.texts[0].set_color(p1.get_color())

    par.xaxis.get_label().set_color(p2.get_color())
    leg.texts[1].set_color(p2.get_color())

    plt.ylabel(tensor_name)

    plt.show()
```

```python
plot_tensor(trial, "CrossEntropyLoss_output_0")
```

```python
# TODO: Deploy your model to an endpoint
jpeg_serializer = sagemaker.serializers.IdentitySerializer("image/jpeg")
json_deserializer = sagemaker.deserializers.JSONDeserializer()
from sagemaker.predictor import Predictor

class ImagePredictor(Predictor):
    def __init__(self, endpoint_name, sagemaker_session):
        super(ImagePredictor, self).__init__(
            endpoint_name,
            sagemaker_session=sagemaker_session,
            serializer=jpeg_serializer,
            deserializer=json_deserializer,
        )
```

```python
# TODO: Run an prediction on the endpoint
from sagemaker.pytorch import PyTorchModel

pytorch_model = PyTorchModel(
    model_data=estimator.model_data, 
    role=role, 
    entry_point='inference.py',
    py_version='py3',
    framework_version='1.8',
    predictor_cls=ImagePredictor
)
```
